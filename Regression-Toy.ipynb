# Regression Problem

### module imports

```python
from src.RBF import RBFNet
from src.ES import ES
import random
import numpy as np
import matplotlib.pyplot as plt
from sklearn import preprocessing
```

## First Example: Sine Wave

### Initialization and plot

```python
X = np.arange(-10, 10, 0.1)
y = np.sin(X)
X = X.reshape(X.shape[0], 1)
y = y.reshape(y.shape[0], 1)
X, y = preprocessing.scale(X), preprocessing.scale(y)
plt.scatter(X.ravel(), y.ravel())
plt.show()
```

### ES initialization and training

```python
es = ES(X=X, y=y, MU=10, LAMBDA=100, ngen=50, circles=10, problem='regression')
my_rbf = es.train()
```

### Plotting result

```python
xs = np.arange(np.min(X), np.max(X), 0.01)
ys = []
for x in xs:
    ys.append(my_rbf.apply_network(x))

plt.scatter(X.ravel(), y.ravel(), label='input data')
plt.plot(list(xs), ys, 'r', label='regressioin result')
plt.legend()
plt.show()
```
